```python
v_cam_set = set()
v_cam_id_list = []

for i in v_lines:
    v_new_cam_id = int(i.split(' ')[0].split('_')[1][1:])  # format as 0000_c05_s171429_f12990_nir.jpg
    v_cam_set.add(v_new_cam_id)
    v_cam_id_list.append(v_new_cam_id)

    

t_cam_set = set()
t_cam_id_list = []

for i in t_lines:
    t_new_cam_id = int(i.split(' ')[0].split('_')[1][1:])
    t_cam_set.add(t_new_cam_id)
    t_cam_id_list.append(t_new_cam_id)
```

### 设置each cam有一个dict

```python
v_cam_dict = {1:{},2:{},3:{},4:{},5:{},6:{},7:{},8:{},9:{}}
t_cam_dict = {1:{},2:{},3:{},4:{},5:{},6:{},7:{},8:{},9:{}}
```

```python
for i in v_lines:
    cam_id = int(i.split(' ')[0].split('_')[1][1:])
    person_id = int(i.split(' ')[0].split('/')[1])
    
    # 如果没有键的话
    if not v_cam_dict.get(cam_id).get(person_id):
        v_cam_dict[cam_id][person_id] = 1
    else:
        v_cam_dict[cam_id][person_id] += 1
```

```python
for i in t_lines:
    cam_id = int(i.split(' ')[0].split('_')[1][1:])
    person_id = int(i.split(' ')[0].split('/')[1])
    
    # 如果没有键的话
    if not t_cam_dict.get(cam_id).get(person_id):
        t_cam_dict[cam_id][person_id] = 1
    else:
        t_cam_dict[cam_id][person_id] += 1
```

```python
v_cam_dict
```

Visible里面的ID数量

```python
count = 0

for outer_key, inner_dict in v_cam_dict.items():
    num_keys = len(inner_dict)
    print(f"子字典 {outer_key} 的键数量: {num_keys}")
    count+=num_keys
print(count)
```

Thermal里面的ID数量

```python
count = 0

for outer_key, inner_dict in t_cam_dict.items():
    num_keys = len(inner_dict)
    print(f"子字典 {outer_key} 的键数量: {num_keys}")
    count+=num_keys
print(count)
```
