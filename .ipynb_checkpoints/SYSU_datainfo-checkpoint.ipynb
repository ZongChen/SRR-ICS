# Train Info

```python
train_vis_file = '/media/deep/SSD/Dataset_ReID_VI/LLCM/idx/train_vis.txt'
train_nir_file = '/media/deep/SSD/Dataset_ReID_VI/LLCM/idx/train_nir.txt'
```

```python
import os


with open(os.path.join(train_vis_file), 'rt') as f:
    v_lines = f.read().splitlines()

with open(os.path.join(train_nir_file), 'rt') as f:
    t_lines = f.read().splitlines()
```

```python
len(v_lines), len(t_lines)
```

```python
int(v_lines[0].split(' ')[0].split('_')[1][1:])
```

```python
v_cam_set = set()
v_cam_id_list = []

for i in v_lines:
    v_new_cam_id = int(i.split(' ')[0].split('_')[1][1:])
    v_cam_set.add(v_new_cam_id)
    v_cam_id_list.append(v_new_cam_id)

t_cam_set = set()
t_cam_id_list = []

for i in t_lines:
    t_new_cam_id = int(i.split(' ')[0].split('_')[1][1:])
    t_cam_set.add(t_new_cam_id)
    t_cam_id_list.append(t_new_cam_id)
```

```python
len(v_cam_id_list), len(t_cam_id_list)
```

```python
v_cam_set, t_cam_set
```

```python
V和T中，每个cam的图片数量
```

```python
v_all_id = 0
for x in range(1, 10):
    count = v_cam_id_list.count(x)
    print(x, v_cam_id_list.count(x))
    v_all_id += count

print('v_all->  ', v_all_id)

t_all_id = 0
for x in range(1, 10):
    count = t_cam_id_list.count(x)
    print(x,t_cam_id_list.count(x))
    t_all_id += count

print('t_all->  ', t_all_id)
```
