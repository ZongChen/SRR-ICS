# 索引，ndarray可以用列表索引

```python
import numpy as np

new_labels = np.array([10, 20, 30, 40])
global_label = [1, 2]  # 可以是列表
labels = new_labels[global_label]  # 合法！返回 array([20, 30])
```

```python
labels
```

# Tensor

```python
import torch
```

```python
# 原始张量 [A, B]
x = torch.tensor([[1, 2], [3, 4]])  # shape (2, 2)

# 沿第0维重复2次，第1维重复1次 -> [2*A, B]
y = x.repeat(2, 1)
# 结果：
# [[1, 2],
#  [3, 4],
#  [1, 2],
#  [3, 4]]
print(y.shape)  # torch.Size([4, 2])
```

```python
def generate_curriculum(cam_list, epoch_per_stage=5):
    curriculum = []
    for stage in range(1, len(cam_list) + 1):
        current_cams = cam_list[:stage]  # 截取前 stage 个摄像头
        curriculum += current_cams * epoch_per_stage
    return curriculum
```
